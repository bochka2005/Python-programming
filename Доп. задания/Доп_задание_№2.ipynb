```python
phone_book = {}

def format_name(name):
    return ' '.join(word.capitalize() for word in name.split())

def format_phone(phone):
    digits = ''.join(filter(str.isdigit, phone))

    if digits.startswith('8'):
        digits = '7' + digits[1:]

    if digits.startswith('7'):
        return '+' + digits

    if len(digits) == 10:
        return '+7' + digits

    return '+' + digits

def add_contact():
    name = input("Введите имя контакта: ")
    phone = input("Введите номер телефона: ")

    formatted_name = format_name(name)
    formatted_phone = format_phone(phone)

    phone_book[formatted_name] = formatted_phone
    print("Контакт успешно добавлен!\n")

def delete_contact():
    name = input("Введите имя контакта для удаления: ")
    formatted_name = format_name(name)

    if formatted_name in phone_book:
        del phone_book[formatted_name]
        print("Контакт успешно удален!\n")
    else:
        print("Контакт не найден!\n")

def show_contacts():
    if not phone_book:
        print("Телефонная книга пуста!\n")
        return

    print("Список контактов:")
    for name, phone in phone_book.items():
        print(f"{name} {phone}")
    print()

def change_phone():
    name = input("Введите имя контакта: ")
    formatted_name = format_name(name)

    if formatted_name in phone_book:
        new_phone = input("Введите новый номер телефона: ")
        formatted_phone = format_phone(new_phone)
        phone_book[formatted_name] = formatted_phone
        print("Номер телефона успешно изменен!\n")
    else:
        print("Контакт не найден!\n")

while True:
    print("Выберите функцию:")
    print("1. Добавить контакт")
    print("2. Удалить контакт")
    print("3. Показать список контактов")
    print("4. Изменить номер")
    print("5. Выход")

    choice = input("Введите номер функции: ")

    if choice == '1':
        add_contact()
    elif choice == '2':
        delete_contact()
    elif choice == '3':
        show_contacts()
    elif choice == '4':
        change_phone()
    elif choice == '5':
        print("До свидания!")
        break
    else:
        print("Неверный выбор! Попробуйте снова.\n")
```
