```
import random
a, count = random.randint(0, 100), 0
while True:
    us_num = input("Ваш выбор:")
    if us_num == "Выход":
      break
    us_num = int(us_num)
    count += 1
    if us_num ==a:
      print(f'ДА! Угадано за {count} попыток')
      break
    print('Число меньше' if us_num < a else 'Число больше')
```
